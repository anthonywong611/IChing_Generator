```python
fangTu = pd.read_json(StringIO(json.dumps(六十四卦)))
```

```python
names = pd.melt(fangTu, ignore_index=False, var_name="外卦", value_name="挂名").reset_index(names="内卦")
names['重卦'] = names["外卦"] + names["内卦"]
```

```python
重卦 = {row['挂名']: {'外卦': row['外卦'],'内卦': row['内卦']} for index, row in names.iterrows()}
```

```python
def 错(
   name: str, 
   cuo: dict = {
      "地": "天",
      "山": "泽",
      "水": "火",
      "风": "雷",
      "雷": "风",
      "火": "水",
      "泽": "山",
      "天": "地"
}) -> str:
   外错 = cuo[重卦[name]['外卦']]
   内错 = cuo[重卦[name]['内卦']]
   return 六十四卦[外错][内错]
```

```python
def 综(
   name: str, 
   zong: dict = {
      "地": "地",
      "山": "雷",
      "水": "水",
      "风": "泽",
      "雷": "山",
      "火": "火",
      "泽": "风",
      "天": "天"
}) -> str:
   外综 = zong[重卦[name]['内卦']]
   内综 = zong[重卦[name]['外卦']]
   return 六十四卦[外综][内综]
```

```python
def 互(
   name: str,
   bagua: dict = {
      "地": "000",
      "山": "100",
      "水": "010",
      "风": "110",
      "雷": "001",
      "火": "101",
      "泽": "011",
      "天": "111"
}) -> str:
   bagua_inverted = {v: k for k, v in bagua.items()}
   外卦, 内卦 = 重卦[name]['外卦'], 重卦[name]['内卦']
   chong_binary = f"{bagua[外卦]}{bagua[内卦]}"
   outer_exch, inner_exch = chong_binary[1:4], chong_binary[2:5]
   外互, 内互 = bagua_inverted[outer_exch], bagua_inverted[inner_exch]
   return 六十四卦[外互][内互]
```

```python
def 错综复杂(name: str) -> List[str]:
   """ (原卦，错卦，综卦，错综卦，互卦，互错挂，互综卦，互错综卦)
   """
   four_sides = [name, 错(name), 综(name), 错(综(name))]
   four_change_sides = list(map(互, four_sides))
   return four_sides + four_change_sides
```

```python
df = pd.DataFrame(
   {gua: 错综复杂(gua) for gua in 重卦.keys()},
   index=['本', '错', '综', '错综', '互', '互错', '互综', '互错综']
).transpose().reset_index()
```
